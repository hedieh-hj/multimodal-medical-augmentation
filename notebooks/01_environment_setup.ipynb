```python
base_path = "../"
folders = [
    "data/raw",
    "data/processed",
    "outputs",
    "src"
]

for f in folders:
    path = os.path.join(base_path, f)
    os.makedirs(path, exist_ok=True)
    print(f"Folder ready: {path}")
```

```python
preprocessing_path = os.path.join(base_path, "src", "preprocessing.py")

if not os.path.exists(preprocessing_path):
    with open(preprocessing_path, "w", encoding="utf-8") as f:
        f.write(
            "# This is the preprocessing module\n\n"
            "import pandas as pd\n"
            "import numpy as np\n\n"
            "def load_csv(path):\n"
            "    \"\"\"Load CSV file\"\"\"\n"
            "    return pd.read_csv(path)\n\n"
            "def normalize_array(arr):\n"
            "    \"\"\"Normalize numpy array\"\"\"\n"
            "    return (arr - np.min(arr)) / (np.max(arr) - np.min(arr))\n"
        )
    print("Preprocessing module created at:", preprocessing_path)
else:
    print("Preprocessing module already exists.")
```
